```python
# [(words, category)]
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[:10]
```

```python
# remove stop words and apply lemmatization
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    else :
        return wordnet.NOUN
```

```python
w="Learning"
pos_tag([w]) 
```

```python
stops = set(stopwords.words('english'))
stops.update(list(string.punctuation))
print(stops) 
print(string.punctuation[:5])
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1])) 
            output_words.append(clean_word.lower())
    return output_words
```

```python
doc=[]
for document,category in documents:
    doc.append((clean_review(document),category))
```

```python
doc[0]
```

```python
training_documents = doc[0:1500]
testing_documents = doc[1500:]
training_documents[0:3]
```

```python
all_words=[]
for doc in training_documents:
    all_words += list(doc[0])
```

```python
freq = nltk.FreqDist(all_words)
```

```python
freq.most_common(10)
```

```python
features = [i[0] for i in freq.most_common(3000)]
features[:5]
```

```python
def get_feature_dict(words):
    current_features={}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data=[(get_feature_dict(doc), category) for doc, category in training_documents]
```

```python
get_feature_dict(training_documents[0][0])
```

```python
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_documents]
```

```python
# Naive Bayes classifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```
